```python
def test_post(session, url, access_token, owner_id, album_id, count=100):

    headers = {
        'Origin': 'https://vkvideo.ru',
        # 'Content-Type': 'application/json'
    }

    data = {
        'v': '5.268',
        'client_id': '52461373',
        'album_id': album_id,
        'count': count,
        'extended': 1,
        'fields': 'is_esia_verified%2Cis_sber_verified%2Cis_tinkoff_verified%2Cphoto_50%2Cverified',
        'offset': 0,
        'owner_id': owner_id,
        'sort_album':0,
        'access_token': access_token
    }

    resp = session.post(url=url, data=data, headers=headers)
    text = resp.text
    headers = resp.headers
    return {'resp': resp.json(), 'headers': headers, 'cookies': resp.cookies}
```

```python
def extract_from_json(idx, owner_id, album_id, json_response):
    res = []

    for i, v in enumerate(json_response['response']['items']):
        # print(v.get('video')['video'])
        if not v['video'].get('title'):
            print('something wrong with', i, v['video'])
            if not failed.get(idx):
                failed[idx] = dict()
            failed[idx][i] = v['video']
            continue
        # print(v['video']['playlist_position'])
        res.append({
            '_album_index': idx,
            '_album_owner_id': owner_id,
            '_album_album_id': album_id,
            'id': v['video'].get('id'),
            'owner_id': v['video'].get('owner_id'),
            'title': v['video'].get('title'),
            'direct_url': v['video'].get('direct_url'),
            'adding_date': v['video'].get('adding_date'),
            'date': v['video'].get('date'),
            'description': v['video'].get('description'),
            'duration': v['video'].get('duration')
        })

    return res
```

# Извлекаем гостей с помощью spacy

```python
df = pd.read_csv('df.csv')
df.head(50)
```

```python
import spacy
nlp = spacy.load("ru_core_news_sm")

def extract_names(text):
    if text:
        doc = nlp(text)
        res = []
        for e in doc.ents:
            if e.label_ == 'PER':
                res.append(e.text)
        return res

guests_from_description = df['description'].fillna('').apply(extract_names)
guests_from_description
```

```python
guests_from_title = df['title'].fillna('').apply(extract_names)
guests_from_title
```

```python
df['guests_from_title'] = guests_from_title
df['guests_from_description'] = guests_from_description
```

```python
d = dict()

for i in list(guests_from_title + guests_from_description):
    if type(i)==list:
        for name in i:
            if ' ' in name:
                print(name)
                try:
                    first_name, last_name = name.split()[:2]
                    d[last_name] = name
                except:
                    pass
```
